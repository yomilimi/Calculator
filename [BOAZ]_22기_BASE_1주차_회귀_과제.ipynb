## 단순 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# 데이터셋 설정

x_train = np.array([1., 2., 3., 4., 5., 6.])
y_train = np.array([9., 16., 23., 30., 33., 50.])

# W,b 초기값 설정
W = 0.0
b = 0.0

# 데이터셋 길이 저장
n_data = len(x_train)

# 학습률 파라미터 설정
epochs = 5000
learning_rate = 0.01

# 학습(경사하강법)
for i in range(epochs):
    hypothesis = x_train * W + b
    cost = np.sum((hypothesis - y_train) ** 2) / n_data
    gradient_w = np.sum((W * x_train - y_train + b) * 2 * x_train) / n_data
    gradient_b = np.sum((W * x_train - y_train + b) * 2) / n_data

    W -= learning_rate * gradient_w
    b -= learning_rate * gradient_b

    if i % 100 == 0:
        print('Epoch ({:10d}/{:10d}) cost: {:10f}, W: {:10f}, b:{:10f}'.format(i, epochs, cost, W, b))

# 결과 출력
print('W: {:10f}'.format(W))
print('b: {:10f}'.format(b))
print('회귀 식 : Y={:10f}x +{:10f}'.format(W,b))
print('result : ',x_train * W + b)
```

## 과제 1
#### 최적의 학습률 찾기

```python
############과제#############

# 학습률(learning_rate)을 바꿔보며 Nan이 나오는 값은 얼마인지, 적절한 학습률 값은 무엇인지에 대해 자유롭게 써주세요.
W = 0.0
b = 0.0
learning_rate = 0.1

for i in range(epochs):
    hypothesis = x_train * W + b
    cost = np.sum((hypothesis - y_train) ** 2) / n_data
    gradient_w = np.sum((W * x_train - y_train + b) * 2 * x_train) / n_data
    gradient_b = np.sum((W * x_train - y_train + b) * 2) / n_data

    W -= learning_rate * gradient_w
    b -= learning_rate * gradient_b

    if i % 100 == 0:
        print('Epoch ({:10d}/{:10d}) cost: {:10f}, W: {:10f}, b:{:10f}'.format(i, epochs, cost, W, b))
```

#### 과제에 대한 답을 적어주세요.

0.01: Epoch ( 4900/ 5000) cost: 7.784127, W: 7.514286, b: 0.533333

0.001: Epoch ( 4900/ 5000) cost: 7.791349, W: 7.469052, b: 0.726989

0.0001: Epoch ( 4900/ 5000) cost: 7.964775, W: 7.287975, b: 1.502208

=> 0.01에서 cost가 가장 작기 때문에 가장 적절한 learning rate이다.

### 보스턴 주택 가격 예측

```python
#보스턴 주택 데이터셋
from sklearn import datasets
housing = datasets.load_boston()

#key 값 확인
housing.keys()
```

```python
# 판다스 데이터 프레임으로 변환
data = pd.DataFrame(housing['data'], columns = housing['feature_names'])
target = pd.DataFrame(housing['target'], columns = ['Target'])

# 데이터셋 크기
print(data.shape)
print(target.shape)
```

```python
# 데이터 프레임 결합
df = pd.concat([data, target], axis=1)
df.head()
```

### 데이터 전처리

#### feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() # MinMaxScaler를 활용한 정규화

df_scaled = df.iloc[:, :-1]   # 마지막 열을 나타내는 -1은 포함하지 않음
scaler.fit(df_scaled)
df_scaled = scaler.transform(df_scaled)

# 스케일링 변환된 값을 데이터 프레임에 반영
df.iloc[:, :-1] = df_scaled[:, :]
df
```

```python
# 학습 데이터와 테스트 데이터 분할
from sklearn.model_selection import train_test_split
x_data=df.loc[:, ['LSTAT', 'RM']]
y_data=df.loc[:, 'Target']
X_train, X_test, y_train, y_test=train_test_split(x_data,
                                                  y_data,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```
